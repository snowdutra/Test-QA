```python
import sys
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver = webdriver.Chrome()
driver.get('https://www.kaggle.com/datasets/jaderz/hospital-beds-management/')
```

```python
botao_cookie = WebDriverWait(driver, 20).until(
	EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'div.sc-ggWZvA.bxFwkO'))
)

botao_cookie[1].click()
```

```python
botao_staff = WebDriverWait(driver, 20).until(
	EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'p.sc-fbQrwq.sc-kqpVfT.jnmBCO.cvjXol'))
)

botao_staff[1].click()
```

```python
botao_coluna = WebDriverWait(driver, 20).until(
	EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'button.MuiButtonBase-root.MuiTab-root.MuiTab-textColorPrimary.sc-bhjLtx.gjRJZI.css-i5k1e3'))
)

botao_coluna[2].click()
```

```python
valores_unicos = WebDriverWait(driver, 20).until(
	EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'p.sc-kNOymR.sc-gckaLG.kagITs.cbWgwV'))
)

ids_unicos = int(valores_unicos[0].text)
nomes_unicos = int(valores_unicos[1].text)
```

```python
spans_nome = WebDriverWait(driver, 20).until(
	EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'span.sc-ipUnzB.sc-iJlEmj.jtxgoS.fYAPvG, span.sc-ipUnzB.sc-iJlEmj.jtxgoS.iILxQZ'))
)

spans_valor = WebDriverWait(driver, 20).until(
	EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'span.sc-ipUnzB.sc-gHWbVC.jtxgoS.jDBNCw, span.sc-ipUnzB.sc-gHWbVC.jtxgoS.YWygX'))
)

cargos = [
	{
		"nome": spans_nome[0].text,
		"valor": int(spans_valor[0].text[:-1])
	},
	{
		"nome": spans_nome[1].text,
		"valor": int(spans_valor[1].text[:-1])
	},
	{
		"nome": spans_nome[2].text,
		"valor": int(spans_valor[2].text[:-1])
	}
]

servicos = [
	{
		"nome": spans_nome[3].text,
		"valor": int(spans_valor[3].text[:-1])
	},
	{
		"nome": spans_nome[4].text,
		"valor": int(spans_valor[4].text[:-1])
	},
	{
		"nome": spans_nome[5].text,
		"valor": int(spans_valor[5].text[:-1])
	}
]
```

```python
print(ids_unicos)

print(nomes_unicos)

print(cargos)

print(servicos)

driver.close()
```
