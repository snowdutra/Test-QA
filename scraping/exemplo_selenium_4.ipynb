```python
import sys
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver = webdriver.Chrome()
driver.get('https://portal.espm.br')
```

```python
link_meus_cursos = WebDriverWait(driver, 20).until(
	EC.presence_of_element_located((By.CSS_SELECTOR, 'a[href="/my-courses"]'))
)

link_meus_cursos.click()
```

```python
link_curso = WebDriverWait(driver, 20).until(
	EC.presence_of_element_located((By.CSS_SELECTOR, 'a[href="/my-courses/student-GTECHSP"]'))
)

link_curso.click()
```

```python
link_professores = WebDriverWait(driver, 20).until(
	EC.presence_of_element_located((By.CSS_SELECTOR, '#simple-tab-1'))
)

link_professores.click()
```

```python
nomes = WebDriverWait(driver, 20).until(
	EC.presence_of_all_elements_located((By.CSS_SELECTOR, '.ProfessorCard_name__lsK_b'))
)
```

```python
emails = WebDriverWait(driver, 20).until(
	EC.presence_of_all_elements_located((By.CSS_SELECTOR, '.ProfessorCard_job__effQA'))
)
```

```python
dados = []

for i in range(len(nomes)):
	nome = nomes[i].text.strip()
	email = emails[i].text.strip()

	dados.append({
		"nome": nome,
		"email": email
	})
```

```python
print(dados)

driver.close()
```
