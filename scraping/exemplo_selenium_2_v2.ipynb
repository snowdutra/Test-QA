```python
import sys
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def extrair_float(texto):
	try:
		i = texto.rindex(' ')
		sem_unidade = texto[:i]

		# Ã€s vezes, esse valor pode iniciar pelo ano...
		i = sem_unidade.find(' ')
		if i >= 0:
			sem_unidade = sem_unidade[(i + 1):]

		sem_virgula = sem_unidade.replace(',', '')

		return float(sem_virgula)
	except:
		return 0
```

```python
driver = webdriver.Chrome()
driver.get('https://ourworldindata.org/explorers/energy')
```

```python
botao_cookie = WebDriverWait(driver, 20).until(
	EC.element_to_be_clickable((By.CSS_SELECTOR, 'button.accept'))
)

# Ã€s vezes o botÃ£o nÃ£o era clicÃ¡vel de verdade logo de primeira ðŸ˜… ...
time.sleep(2)

botao_cookie.click()
```

```python
botoes_tabela = WebDriverWait(driver, 20).until(
	EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'button[data-track-note="chart_click_Table"]'))
)

botoes_tabela[0].click()
```

```python
input = WebDriverWait(driver, 20).until(
	EC.presence_of_element_located((By.CSS_SELECTOR, 'input[data-track-note="data_table_search"]'))
)

input.send_keys('South America')
```

```python
corpo_tabela = WebDriverWait(driver, 20).until(
	EC.presence_of_element_located((By.CSS_SELECTOR, 'div.DataTable table > tbody'))
)
```

```python
linhas = corpo_tabela.find_elements(By.TAG_NAME, 'tr')
```

```python
dados = []

for linha in linhas:
	colunas = linha.find_elements(By.TAG_NAME, 'td')

	nome = colunas[0].text
	valor1965 = colunas[1].text
	valor2024 = colunas[2].text

	dados.append({
		'nome': nome,
		'valor1965': extrair_float(valor1965),
		'valor2024': extrair_float(valor2024)
	})
```

```python

print(dados)

driver.close()
```
