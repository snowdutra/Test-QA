```python
import sys
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def extrair_float(texto):
	try:
		i = texto.rindex(' ')
		sem_unidade = texto[:i]


		i = sem_unidade.find(' ')
		if i >= 0:
			sem_unidade = sem_unidade[(i + 1):]

		sem_virgula = sem_unidade.replace(',', '')

		return float(sem_virgula)
	except:
		return 0

driver = webdriver.Chrome()
driver.get('https://www.imdb.com/pt/chart/moviemeter')
```

```python
botao_exibdetalhada = WebDriverWait(driver, 20).until(
	EC.element_to_be_clickable((By.CSS_SELECTOR, '#list-view-option-detailed'))
)

botao_exibdetalhada.click()
```

```python
elementos_filmes = WebDriverWait(driver, 20).until(
    EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'li.ipc-metadata-list-summary-item'))
)

time.sleep(4)
```

```python
elementos_filmess = WebDriverWait(driver, 20).until(
    EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'div.ipc-html-content-inner-div'))
)

time.sleep(4)
```

```python
ano_filmes = WebDriverWait(driver, 20).until(
    EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'div.sc-caa65599-6.hhUutV.dli-title-metadata'))
)

time.sleep(4)
```

```python
for i, filme in enumerate(elementos_filmes, start=1):
    try:
        titulo_elemento = filme.find_element(By.CSS_SELECTOR, 'h3.ipc-title__text')
        titulo = titulo_elemento.text
        ano_elemento = ano_filmes[i - 1].find_element(By.CSS_SELECTOR, 'span:nth-child(1)')
        ano = ano_elemento.text
        print(f"{i} - {titulo} ({ano})")
    except:
        print(f"{i} - Não foi possível capturar o título")
```
